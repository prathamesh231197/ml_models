**The data consists of biomechanics features of the patients according to their current conditions. Each patient is 
represented in the data set by six biomechanics attributes derived from the shape and orientation of the condition to their body part**. 

```python
#Importing all the necessary libraries fro EDA and Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the csv files to Dataframe
normal = pd.read_csv("Normal.csv")
type_h = pd.read_csv("Type_H.csv")
type_s = pd.read_csv("Type_S.csv")
print(normal,"\n", type_h,"\n", type_s)


```

```python
#Printing the shape of dataframe
shape_normal = normal.shape
shape_type_h = type_h.shape
shape_type_s = type_s.shape
print(f"The shape of 3 dataframes is as follows,\n 1) Normal has {shape_normal[0]} rows & {shape_normal[1]} columns.\n 2) Type_H has {shape_type_h[0]} rows & {shape_type_h[1]} columns.\n 3) Type_S has {shape_type_s[0]} rows & {shape_type_s[1]} columns.")
```

```python
#Printing columns of the dataframe
print("Normal dataframe has the following columns",normal.columns.tolist())
print("Type_h dataframe has the following columns",type_h.columns.tolist())
print("Type_s dataframe has the following columns",type_s.columns.tolist())
```

#### After observing the column names of all 3 dataframes, it is clear that all the dataframes have same column names.

```python
#Printing datatypes of the dataframes.
print("Normal dataframe's datatypes are as follows")
print(normal.dtypes)
print("*"*30)
print("Type_h dataframe's datatypes are as follows")
print(type_h.dtypes)
print("*"*30)
print("Type_s dataframe's datatypes are as follows")
print(type_s.dtypes)
```

```python
#Variation in "Cass" column of dataframes.
variation1 = normal["Class"].unique()
variation2 = type_h["Class"].unique()
variation3 = type_s["Class"].unique()

print(f"There are {len(variation1)} variations of 'Normal' in first dataframe which are '{variation1[0]}' & '{variation1[1]}'.")
print(f"There are {len(variation2)} variations of 'Type_h' in second dataframe which are '{variation2[0]}' & '{variation2[1]}'.")
print(f"There are {len(variation3)} variations of 'Type_s' in third dataframe which are '{variation3[0]}' & '{variation3[1]}'.")
```

```python
#Unifying all the variations in ‘Class’ feature for all the 3 DataFrames.
#I will be using the above variation list created to replace the values.
normal["Class"] = normal["Class"].replace(variation1,"normal")
type_h["Class"] = type_h["Class"].replace(variation2,"type_h")
type_s["Class"] = type_s["Class"].replace(variation3,"type_s")

print("The values in Class column of all dataframes has been harmonized to 'normal', 'type_h' & 'type_s' respectively.")
```

```python
#Combining all the 3 DataFrames to form a single DataFrame
df = pd.concat([normal,type_h,type_s],ignore_index=True)
shape_comb=df.shape
print(f"After combining all three dataframes we have {shape_comb[0]} rows and {shape_comb[1]} columns.")
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score,precision_score, recall_score, f1_score, confusion_matrix
```

```python
df.head()
```

```python
#Split data into X and Y.
#Target feature
y = df["Class"]
#Independent feature
X = df.drop('Class', axis=1)
```

```python
#Split data into train and test with 80:20 proportion
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.20, random_state = 42)
```

```python
#Building KNN Model
m1 = KNeighborsClassifier()
```

```python
#Train a Supervised Learning Classification base model using KNN classifier
m1.fit(X_train, y_train)
```

```python
#Accuracy score
y_test_predicted = m1.predict(X_test)
accuracy = m1.score(X_test, y_test)
print("Test Accuracy is {:.2f}".format(accuracy))
```

```python
#Creating confusion matrix for observing the effectiveness of model predictions
cm = metrics.confusion_matrix(y_test, y_test_predicted)
```

```python
df_cm = pd.DataFrame(cm, index=[i for i in ["normal",'type_h',"type_s"]],
                     columns = [i for i in ["predict_normal","predict_type_h","predict_type_s"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True);
```

```python
#Performance Matrix for Test data
test_report = classification_report(y_test, y_test_predicted)
print(test_report)
```

```python
y_train_predicted = m1.predict(X_train)
train_report = classification_report(y_train, y_train_predicted)
print(train_report)
```

```python
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted))
print("Test Accuracy:", accuracy_score(y_test, y_test_predicted))
print("Precision:", precision_score(y_test, y_test_predicted, average='weighted'))
print("Recall:", recall_score(y_test, y_test_predicted, average='weighted'))
print("F1-Score:", f1_score(y_test, y_test_predicted, average='weighted'))
print("Confusion Matrix:\n",cm)
```

```python
#Performance improvement
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': range(1, 21)
    }
grid_search = GridSearchCV(m1, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_k = grid_search.best_params_
print("Best Hyperparameters:", best_k)
```

```python
best_m1 = KNeighborsClassifier(n_neighbors=12)
best_m1.fit(X_train,y_train)
y_test_predicted_best = best_m1.predict(X_test)
y_train_predicted_best = best_m1.predict(X_train)
print("Performance Metrics for the Best KNN Model:")
print(classification_report(y_test, y_test_predicted_best))
```

```python
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted_best))
print("Test Accuracy:", accuracy_score(y_test, y_test_predicted_best))
print("Precision:", precision_score(y_test, y_test_predicted_best, average='weighted'))
print("Recall:", recall_score(y_test, y_test_predicted_best, average='weighted'))
print("F1-Score:", f1_score(y_test, y_test_predicted_best, average='weighted'))
print("Confusion Matrix:\n",metrics.confusion_matrix(y_test,  y_test_predicted_best))
```
